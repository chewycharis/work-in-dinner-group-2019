**Import Modules and Trajectory**

```python
import numpy as np 
traj=np.load('../traj_unperturb.npy')[:1000, 1::3, :]; traj_flat=np.vstack(traj)
#traj=np.load('../data/erik/muller_brown_traj.npy')[:,:,:2]; traj_flat=np.vstack(traj)
print(np.shape(traj))
```

**Define States**

*code taken from John Strahn*

```python
def ellipseA(x):
    cntr_X=-.5
    cntr_y=1.5
    a=-6.5
    b=11.
    c=-6.5
    r=.3
    ans=np.zeros(len(x))
    for i in range(0,len(x)):
        ans[i]=a*(x[i,0]-cntr_X)**2+b*(x[i,0]-cntr_X)*(x[i,1]-cntr_y)+c*(x[i,1]-cntr_y)**2+r>0
    return ans

def ellipseB(x):
    ans=np.zeros(len(x))
    for i in range(0,len(x)):
        ans[i]=-5.0*(x[i,1]-0.02)**2 - (x[i,0]-0.6)**2 + 0.2>0
    return ans
```

```python
stateA=ellipseA(traj_flat);
stateB=ellipseB(traj_flat);
```

**Build Diffusion Map Basis**

*change data format for using pyedgar*

```python
n_traj=1000; n_frame_per_traj=7
stateA=stateA.reshape((n_traj,n_frame_per_traj)); 
stateB=stateB.reshape((n_traj,n_frame_per_traj)); 
trajs = [traj_i for traj_i in traj]
stateA = [A_i for A_i in stateA]
stateB = [B_i for B_i in stateB]
in_domain = [1. - B_i - A_i for (A_i, B_i) in zip(stateA, stateB)] #if in either, get 0, if in neither, get 1 
```

```python
import pyedgar
diff_atlas = pyedgar.basis.DiffusionAtlas.from_sklearn(alpha=0, k=64, bandwidth_type='-1/d', epsilon='bgh_generous')
diff_atlas.fit(trajs)
basis, evals = diff_atlas.make_dirichlet_basis(300, in_domain=in_domain, return_evals=True)
guess = diff_atlas.make_FK_soln(stateB, in_domain=in_domain)
```

*plot guess function and basis*

```python
import matplotlib.pyplot as plt 
flat_basis = np.vstack(basis)
flat_guess = np.hstack(guess)
fig, axis = plt.subplots(1,2,figsize=(12,4))
axis[0].scatter(traj_flat[:,0], traj_flat[:,1], c= flat_basis[:,0]); axis[0].set_title("first basis")
axis[1].scatter(traj_flat[:,0], traj_flat[:,1], c= flat_guess); axis[1].set_title("guess function")
```

**Compute Committor** 

```python
import scipy 

def compute_generator(x,y=None):
    if y is None:
        x_init=np.vstack(x[:,:-1,:])
        x_end=np.vstack(x[:,1:,:])
        L = x_init.T .dot(x_end) - x_init.T .dot(x_init)
        L *= 1/(len(x_init))**2
        return L
    else:
        x_init=np.vstack(x[:,:-1,:])
        y_init=np.hstack(y[:,:-1])
        y_end =np.hstack(y[:,1:]) 
        L = x_init.T .dot(y_end) - x_init.T .dot(y_init)
        L *= 1/(len(x_init))**2
        return L


def compute_committor(basis, guess, in_domain, grad_basis=None, force=None,coeff_q0=None):
    fLf=compute_generator(np.asarray(basis))
    fLr=compute_generator(np.asarray(basis),np.asarray(guess))
    soln = scipy.linalg.solve(fLf ,-1*fLr )
    diff=np.vstack(np.asarray(basis)) .dot(soln) 
    diff = diff.reshape((1000,7))
    q= diff + np.asarray(guess)
    q[q<0]=0; q[q>1]=1 #apply periodic boundary condition 
    q=np.hstack(q)
    return q
```

```python
g = pyedgar.galerkin.compute_committor(basis, guess, lag=1)
flat_g=np.hstack(g)
q=compute_committor(basis,guess,in_domain)
print(flat_g)
```

```python
plt.clf()
fig, axis = plt.subplots(1,2,figsize=(10,4))
traj_plot=traj[:,:-1,:]; traj_plot=np.vstack(traj_plot)
color_map=axis[0].scatter(traj_flat[:,0], traj_flat[:,1], c=q ); axis[0].set_title("calculated committor from my code")
#axis[1].scatter(traj_flat[:,0], traj_flat[:,1], c=flat_g); axis[1].set_title("calculated committor from pyedgar")
axis[1].scatter(traj_flat[:,0], traj_flat[:,1], c=flat_guess); axis[1].set_title("guess function")
fig.tight_layout()
axis[1].set_xlim(-2.5,1.25); axis[1].set_ylim(-0.25,2.25)
axis[0].set_xlim(-2.5,1.25); axis[0].set_ylim(-0.25,2.25)
plt.set_cmap("Spectral_r")
fig.colorbar(color_map)
```
