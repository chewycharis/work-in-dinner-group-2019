```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt 
```

```python
u=np.random.uniform(0, 1, size=10000)
pi_hat=1/(2*u)**2
pi_hat_plot=pi_hat[np.where(pi_hat<=1)]
```

```python
x=np.arange(0,1,0.05)
pi=1/(2*np.sqrt(x[1:]))
```

```python
fig, ax = plt.subplots(2,1)
ax[0].hist(pi_hat_plot)
ax[0].set_title('pi_hat')
ax[1].plot(x[1:],pi)
ax[1].set_title('pi')
fig.tight_layout()
```
